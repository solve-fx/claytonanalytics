```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("./datasets/boston.csv")

x = np.array(data['RM'])
y = np.array(data['MV'])
```

```python
fig = px.scatter(x= x, y=y)
fig.show()
```

```python
# Column of 1s for intercept
ones = np.ones(x.shape[0]).reshape(x.shape[0], 1)

# Identity Matrix
X = x.reshape(x.shape[0], 1)
X = np.concatenate((ones, X), axis = 1)

# Identity matrix transposed
XT = X.T

# Dot product of transposed identity matrix and the identity matrix
XTX = np.dot(XT, X)

# Inverse of XTX
XTX_inv = np.linalg.inv(XTX)

# Dot product of transposed identity matrix and y
XTy = np.dot(XT,y)

# Dot product of inversed XTX and XTy
beta_hat = np.dot(XTX_inv, XTy)
```

```python
beta_hat
```

```python
yhat = X.dot(beta_hat)
```

```python
fig1 = px.scatter(x=x, y=y)

fig2 = px.line(x=x, y=yhat)
fig2['data'][0]['line']['color']="red"

fig3 = go.Figure(data=fig1.data + fig2.data)
fig3.show()
```
